```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
data=np.loadtxt('resultadosjupiterchidos.txt')
```

```python
fig=plt.figure()

fig = plt.figure()
ax= Axes3D(fig)
x1=data[:,0:1]
y1=data[:,1:2]
z1=data[:,2:3]


ax.scatter(x1,y1,z1, c='r', marker='o')

plt.show()
```

```python
fig=plt.figure()

fig = plt.figure()
ax= Axes3D(fig)
X= np.arange(-4, 4, 0.25)
Y= np.arange(-4, 4, 0.25)
X, Y =np.meshgrid(X,Y)
R=np.sqrt(X**2 + Y**2)
Z=np.sin(R)

ax.plot_surface(X,Y,Z, rstride=1,cstride=1, cmap='hot')
```
