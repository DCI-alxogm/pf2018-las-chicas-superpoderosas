```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
datos=np.loadtxt('1.txt')
```

```python
fig=plt.figure()

fig = plt.figure()
ax= Axes3D(fig)
x=datos[:,0:1]
y=datos[:,1:2]
z=datos[:,2:3]


plt.xlim(1e8,8e12)
plt.ylim(0,1.5e11) # Limitamos los valores del eje x 
ax.set_zlim(0, 5e11)

plt.title('1')
ax.scatter(x,y,z, c='r', marker='o')
plt.show()
```
