```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python

data=np.loadtxt('iniciales.txt')
```

```python
fig=plt.figure()

fig = plt.figure()
ax= Axes3D(fig)
x=data[:,0:1]
y=data[:,1:2]
z=data[:,2:3]

ax.scatter(x,y,z, c='r', marker='o')
plt.title('Posiciones iniciales')
plt.show()
```
